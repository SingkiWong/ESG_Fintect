```python
df=pd.read_csv('raw_news_collection.csv',encoding='gbk',header=0)
```

```python
df['title'].head()
```

```python
df.head()
```

```python
keywords = ['环保', '污染', '处罚', '违规', '债务', '违约', '纠纷', '毁约', '拖欠', '安全性', '质量','投诉','被罚','仲裁']
def clean_ESG(text):
    num=0
    for k in keywords:
        if (k in text):
            num -= 5
    return num
```

```python
rr=df['title'].apply(lambda x: clean_ESG(x))
```

```python
df['select_score']=rr
```

```python
df.head()
```

```python
df_test=df
```

```python
df_test[df_test['select_score']<0].reset_index().to_excel('select.xlsx',index=None)
```
